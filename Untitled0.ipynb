```python
import os, glob, shutil, random, warnings
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import cv2
warnings.filterwarnings('ignore')

print("✅ Setup done")
```

```python
import kagglehub

path = kagglehub.dataset_download("meowmeowmeowmeowmeow/gtsrb-german-traffic-sign")
print("📁 Dataset path:", path)
```

```python
def find_dir(base, name):
    for root, dirs, _ in os.walk(base):
        for d in dirs:
            if d.lower() == name.lower():
                return os.path.join(root, d)
    return None

def find_file(base, name):
    for root, _, files in os.walk(base):
        for f in files:
            if f.lower() == name.lower():
                return os.path.join(root, f)
    return None

TRAIN_ROOT = find_dir(path, 'Train')
TEST_ROOT  = find_dir(path, 'Test')
TEST_CSV   = find_file(path, 'Test.csv')

print(TRAIN_ROOT, TEST_ROOT, TEST_CSV)
```

```python
print("Classes:", len(os.listdir(TRAIN_ROOT)))
```

```python
BASE = "/content/gtsrb_yolo"

for split in ["train", "val", "test"]:
    os.makedirs(f"{BASE}/{split}/images", exist_ok=True)
    os.makedirs(f"{BASE}/{split}/labels", exist_ok=True)

# Train → split
for cid in range(43):
    class_dir = os.path.join(TRAIN_ROOT, str(cid))
    imgs = glob.glob(os.path.join(class_dir, "*.ppm")) + \
           glob.glob(os.path.join(class_dir, "*.png"))

    random.shuffle(imgs)
    split_idx = int(0.8 * len(imgs))

    for i, img_path in enumerate(imgs):
        img = cv2.imread(img_path)
        h, w = img.shape[:2]

        # full image bbox (classification → detection trick)
        label = f"{cid} 0.5 0.5 1.0 1.0"

        split = "train" if i < split_idx else "val"

        fname = os.path.basename(img_path)
        new_img = f"{BASE}/{split}/images/{cid}_{fname}"
        new_lbl = f"{BASE}/{split}/labels/{cid}_{fname.replace('.ppm','.txt').replace('.png','.txt')}"

        shutil.copy(img_path, new_img)

        with open(new_lbl, "w") as f:
            f.write(label)

# Test images
for img_path in glob.glob(os.path.join(TEST_ROOT, "*.ppm")):
    fname = os.path.basename(img_path)
    shutil.copy(img_path, f"{BASE}/test/images/{fname}")

print("✅ YOLO dataset ready")
```

```python
names = [f"class_{i}" for i in range(43)]

yaml_content = f"""
path: {BASE}
train: train/images
val: val/images
test: test/images

nc: 43
names: {names}
"""

YAML_PATH = f"{BASE}/dataset.yaml"

with open(YAML_PATH, "w") as f:
    f.write(yaml_content)

print("✅ YAML created")
print(yaml_content)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data=YAML_PATH,
    epochs=5,
    imgsz=640,
    batch=32,
    device=0,
    project="/content/gtsrb_runs",
    name="yolo11_gtsrb",
    exist_ok=True
)
```

```python
model = YOLO("/content/gtsrb_runs/yolo11_gtsrb/weights/best.pt")

metrics = model.val(data=YAML_PATH)

print("mAP50:", metrics.box.map50)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```
